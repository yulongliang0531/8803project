```python
import matplotlib.pyplot as plt
import numpy as np

def init_func(x):
    y = 1600 - 3200 * (x ** 2) + 1600 * (x ** 4)
    return y

n = 100
base_x = np.linspace(0, 1, n)
base_y = np.zeros(n)
for i in range(n):
    base_y[i] = init_func(base_x[i])

def plot_gpr_samples(gpr_model, n_samples, ax):
    """Plot samples drawn from the Gaussian process model.

    If the Gaussian process model is not trained then the drawn samples are
    drawn from the prior distribution. Otherwise, the samples are drawn from
    the posterior distribution. Be aware that a sample here corresponds to a
    function.

    Parameters
    ----------
    gpr_model : `GaussianProcessRegressor`
        A :class:`~sklearn.gaussian_process.GaussianProcessRegressor` model.
    n_samples : int
        The number of samples to draw from the Gaussian process distribution.
    ax : matplotlib axis
        The matplotlib axis where to plot the samples.
    """
    n = 100
    x = np.linspace(0, 1, n)
    X = x.reshape(-1, 1)

    y_mean, y_std = gpr_model.predict(X, return_std=True)
    y_samples = gpr_model.sample_y(X, n_samples)
    #print(len(y_samples.T[0]))
    for idx, single_prior in enumerate(y_samples.T):
        # adjust to fix the point [1,0]
        single_prior = single_prior - single_prior[n-1]
        # add base line
        single_prior = single_prior + base_y
        # normalized pressure profile
        single_prior = single_prior / single_prior[0]
        ax.plot(
            x,
            single_prior,
            linestyle="-",
            alpha=0.7,
            color = "blue",
        )
    ax.set_xlabel(r'$\rho$')
    ax.set_ylabel(r'$p(\rho)$')
```

```python
n_samples = 1000
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 2000.0 * RBF(length_scale=0.1, length_scale_bounds="fixed")
gpr = GaussianProcessRegressor(kernel=kernel, random_state=0)

fig, ax = plt.subplots()

# plot prior
plot_gpr_samples(gpr, n_samples=n_samples, ax=ax)
base_y = base_y / base_y[0]
ax.plot(base_x, base_y, color = "red", label = "base", linewidth=2)
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def init_func(x):
    y = 1600 - 3200 * (x ** 2) + 1600 * (x ** 4)
    return y

n = 100
base_x = np.linspace(0, 1, n)
base_y = np.zeros(n)
for i in range(n):
    base_y[i] = init_func(base_x[i])

def plot_gpr_samples(gpr_model, n_samples, ax):
    """Plot samples drawn from the Gaussian process model.

    If the Gaussian process model is not trained then the drawn samples are
    drawn from the prior distribution. Otherwise, the samples are drawn from
    the posterior distribution. Be aware that a sample here corresponds to a
    function.

    Parameters
    ----------
    gpr_model : `GaussianProcessRegressor`
        A :class:`~sklearn.gaussian_process.GaussianProcessRegressor` model.
    n_samples : int
        The number of samples to draw from the Gaussian process distribution.
    ax : matplotlib axis
        The matplotlib axis where to plot the samples.
    """
    n = 100
    x = np.linspace(0, 1, n)
    X = x.reshape(-1, 1)
    store = []

    y_mean, y_std = gpr_model.predict(X, return_std=True)
    y_samples = gpr_model.sample_y(X, n_samples)
    #print(len(y_samples.T[0]))
    for idx, single_prior in enumerate(y_samples.T):
        # adjust to fix the point [1,0]
        single_prior = single_prior - single_prior[n-1]
        # add base line
        single_prior = single_prior + base_y
        # store the line
        store.append(single_prior)
        # normalized pressure profile
        #single_prior = single_prior / single_prior[0]
        ax.plot(
            x,
            single_prior,
            linestyle="-",
            alpha=0.7,
            color = "blue",
        )
    ax.set_xlabel(r'$\rho$')
    ax.set_ylabel(r'$p(\rho)$')
    return store
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 2000.0 * RBF(length_scale=0.1, length_scale_bounds="fixed")
gpr = GaussianProcessRegressor(kernel=kernel, random_state=0)

fig, ax = plt.subplots()

# plot prior
store = plot_gpr_samples(gpr, n_samples=n_samples, ax=ax)
# base_y = base_y / base_y[0]
ax.plot(base_x, base_y, color = "red", label = "base", linewidth=2)
plt.legend()
```

```python
def even_polyfit(x, y, deg):
    # Make degree even
    if deg % 2 != 0:
        deg += 1

    # Calculate the Vandermonde matrix for even-powered terms
    vand_matrix = np.vander(x, deg+1)
    vand_matrix = vand_matrix[:, ::2]

    # Calculate the coefficients using least squares
    coeffs = np.linalg.lstsq(vand_matrix, y, rcond=None)[0]

    return coeffs

# Sample data points
x = np.array([-2, -1, 0, 1, 2])
y = np.array([4, 1, 2, 5, 10])

# Degree of polynomial
deg = 4

# Fit a polynomial with only even-degree terms to the sample data
coeffs = even_polyfit(base_x, base_y, deg)

print("Coefficients:", coeffs)
```

```python
ce1 = []
ce2 = []
ce3 = []
for i in range(len(store)):
    coeffs = even_polyfit(base_x, store[i], 4)
    ce1.append(coeffs[0])
    ce2.append(coeffs[1])
    ce3.append(coeffs[2])
```
